## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')
df['Date'] = pd.to_datetime(df['Date'])

df_2023 = df[df['Date'].dt.year == 2023]

df_unleaded = df_2023[df_2023['Fuel Type'].str.startswith('Regular')]
df_prices = df_unleaded[['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound','St. Catharine\'s']]


plt.figure(figsize=(10, 6))
df_prices.mean().sort_values().plot(kind='bar', color='red')
plt.title('Avg Regular Unleaded Gas in 2023 (by cities)')
plt.xlabel('City')
plt.ylabel('Avg Price (in cents)')
plt.show()


```

## Question 3

Plot a histogram of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023, using 15 bins.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')
df['Date'] = pd.to_datetime(df['Date'])
df_2023_unleaded = df[(df['Date'].dt.year == 2023) & (df['Fuel Type'] == 'Regular Unleaded Gasoline')]
west = df_2023_unleaded['Toronto West/Ouest']

plt.figure(figsize=(10, 6))
plt.hist(west, bins=15, color='red', edgecolor='black')
plt.title('Regular Unleaded Gas Price in West Toronto for 2023')
plt.xlabel('gasoline price')
plt.ylabel('Frequency')
plt.show()

```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')
df['Date'] = pd.to_datetime(df['Date'])
df_2023_unleaded = df[(df['Date'].dt.year == 2023) & (df['Fuel Type'] == 'Regular Unleaded Gasoline')]
west = df_2023_unleaded['Toronto West/Ouest']

fig, ax = plt.subplots(figsize=(10, 6))

ax.ecdf(west, complementary=False, orientation='vertical', marker='.', linewidth=1, color='blue', label='ECDF')

ax.set_title('ECDF of Regular Unleaded Gas Prices in Toronto West/Ouest for 2023')
ax.set_xlabel('Gasoline Price (in cents)')
ax.set_ylabel('ECDF')
ax.legend()
plt.grid(True)
plt.show()

```

## Question 5

Plot a boxplot to display the distribution of Regular Unleaded Gasoline prices across different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") from 2021 onwards (including 2021).


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')
df['Date'] = pd.to_datetime(df['Date'])

df_2023 = df[df['Date'].dt.year >= 2023]

df_unleaded = df_2023[df_2023['Fuel Type'].str.startswith('Regular')]
df_prices = df_unleaded[['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound','St. Catharine\'s']]

plt.figure(figsize=(25, 6))
df_prices.boxplot(grid=True, vert=True)
plt.title('Boxplot of Regular Unleaded Gasoline 2021+')
plt.xlabel('Cities')
plt.ylabel('Gasoline Price')
plt.show()
```
